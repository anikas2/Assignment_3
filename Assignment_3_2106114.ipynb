```python
## RUNNING THE SIMULATION OF THE GAME FOR 10 PIECES OF PAPER IN THE HAT

import random

def simulate_hat_process():
    # initializing the hat with numbers 1 through 10
    hat = list(range(1, 11))

    # initial state of the hat
    print("Initial hat:", hat)

    while len(hat) > 1:
        # draw two pieces of paper at random (without replacement)
        draw1 = random.choice(hat)
        hat.remove(draw1)
        draw2 = random.choice(hat)
        hat.remove(draw2)

        # calculate the difference between these two drawn numbers
        difference = abs(draw1 - draw2)

        # add in this difference to the hat
        hat.append(difference)

        # current state of the hat
        print(f"Draw: {draw1}, {draw2} | Difference: {difference} | Current hat: {hat}")

    # final remaining piece of paper, i.e. the result
    final_result = hat[0]

    return final_result

# run the simulation
result = simulate_hat_process()
print("The final result is:", result)
```

```python
## PLOTTING THE DISTRIBUTION OF RESULTS FOR 100 TRIALS OF 10 PIECES OF PAPER IN THE HAT

import random
import matplotlib.pyplot as plt

# the same process as previously
def simulate_hat_process():
    hat = list(range(1, 11))
    while len(hat) > 1:
        draw1 = random.choice(hat)
        hat.remove(draw1)
        draw2 = random.choice(hat)
        hat.remove(draw2)
        difference = abs(draw1 - draw2)
        hat.append(difference)
    return hat[0]

# simulate the game 100 times
results = [simulate_hat_process() for _ in range(100)]

# count the frequency of each result
result_freq = {x:results.count(x) for x in results}

# create the bar chart
plt.bar(result_freq.keys(), result_freq.values(), color='purple')
plt.xlabel('Final Number, y_N')
plt.ylabel('Frequency that this result is obtained in 100 trials')
plt.title('Frequency of Final Numbers obtained, y_N, in 100 Trials')
plt.xticks(list(result_freq.keys()))
plt.show()
```
